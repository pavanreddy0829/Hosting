## NUMPY ARRAY

```python
"""
NumPy is used to work with arrays. The array object in NumPy is called ndarray.

We can create a NumPy ndarray object by using the array() function.
"""
```

```python
# numpy.array

#passing list 
arr = np.array([1, 2, 3, 4, 5])

print(arr)

print(type(arr))
print("-------------------------------------")
#passing tuple
arr1 = np.array((1, 2, 3, 4, 5))

print(arr1)
```

```python
"""
0-D Arrays
0-D arrays, or Scalars, are the elements in an array. Each value in an array is a 0-D array."""
arr = np.array(42)

print(arr)


print("\n")

"""
1-D Arrays
An array that has 0-D arrays as its elements is called uni-dimensional or 1-D array.

These are the most common and basic arrays."""

arr1 = np.array([1, 2, 3, 4, 5])

print(arr1)


print("\n")

"""
2-D Arrays
An array that has 1-D arrays as its elements is called a 2-D array.

These are often used to represent matrix or 2nd order tensors."""


arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print(arr2)


print("\n")


"""
3-D arrays
An array that has 2-D arrays (matrices) as its elements is called 3-D array.

These are often used to represent a 3rd order tensor"""

arr3 = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr3)


print("\n")

"""
Higher Dimensional Arrays
An array can have any number of dimensions.

When the array is created, you can define the number of dimensions by using the ndmin argument."""


arr4 = np.array([1, 2, 3, 4], ndmin=5)

print(arr4)
print('number of dimensions :', arr4.ndim)
```

## Linspace

```python
"""
The numpy.linspace() function returns number spaces evenly w.r.t interval. Similar to numpy.arrange() function but instead of step it uses sample number. """

print(np.linspace(2.0, 3.0, num=5))

print(np.linspace(2.0, 3.0, num=5, endpoint=False))

print(np.linspace(2.0, 3.0, num=5, retstep=True))
```

```python
np.linspace(start = 0, stop = 100, num = 5)
```

## RANDOM

numpy.random() in Python
The random is a module present in the NumPy library. This module contains the functions which are used for generating random numbers. This module contains some simple random data generation methods, some permutation and distribution functions, and random generator functions.

```python
a=np.random.rand(5,2)  
print(a)  
```

```python
b=np.random.randn()
print(b)
```

```python
a=np.random.randint(3, size=10)  
print(a)
```

```python
out_arr = np.random.random(size = 3)
print ("Output 1D Array filled with random floats : \n", out_arr) 
```

```python
out_arr = np.random.random(size =(2, 4))
print ("Output 2D Array filled with random floats :\n ", out_arr) 
```

```python
out_arr = np.random.random((2, 3, 2))
print ("Output 3D Array filled with random floats : \n", out_arr) 
```
